```python
import tensorflow as tf
import tensorflow_federated as tff
import numpy as np
import collections

def get_dataset(dataset: pd.DataFrame):
    return tf.data.Dataset.from_tensor_slices(dict(dataset))

def get_datasets_per_class(dataset: pd.DataFrame):
    datasets_per_class = {}
    by_class = dataset.groupby("Attack Type")
    for groups, data in by_class:
        datasets_per_class[groups] = {
            "data": tf.data.Dataset.from_tensor_slices(dict(data)),
            "size": len(data.index)
        }
    return datasets_per_class

# def get_empty_dataset(ds_):
#     # # Define an empty generator function
#     # def empty_generator(element_spec):
#     #     for _ in tf.data.Dataset.from_tensor_slices([]).element_spec:
#     #         yield tf.constant([], dtype=_.dtype)

#     # # Create an empty dataset using the generator function
#     # return tf.data.Dataset.from_generator(empty_generator, output_types=original_ds.element_spec)
#     return tf.data.Dataset.from_tensor_slices(tf.reshape((), (0, len())))
```

```python
NUM_CLIENTS = 10
NUM_EPOCHS = 5
BATCH_SIZE = 20
SHUFFLE_BUFFER = 100
PREFETCH_BUFFER = 10

def preprocess(dataset):
    def batch_format_fn(element):
        return element
    
    return dataset.repeat(NUM_EPOCHS).shuffle(SHUFFLE_BUFFER, seed=1).batch(
        BATCH_SIZE).map(batch_format_fn).prefetch(PREFETCH_BUFFER)

# split_method -> ['naive', 'per_class']
def get_federated_clients(dataset: pd.DataFrame, split_method: str):
    c_ids = np.arange(NUM_CLIENTS)
    federated_clients = collections.OrderedDict()
    if split_method == "naive":
        ds = get_dataset(dataset)
        num_examples = len(dataset.index)
        num_examples_per_client = num_examples // NUM_CLIENTS
        for c_id in c_ids:
            client_data = ds.skip(num_examples_per_client * c_id).take(num_examples_per_client)
            federated_clients[c_id] = client_data
    elif split_method == "per_class":
        datasets_per_class = get_datasets_per_class(dataset)
        for c_id in c_ids:
            for d in datasets_per_class.values():
                num_examples = d["size"]
                num_examples_per_client = num_examples // NUM_CLIENTS
                data = d["data"].skip(num_examples_per_client * c_id).take(num_examples_per_client)
                if federated_clients.get(c_id) is None:
                    federated_clients[c_id] = data
                else:
                    federated_clients[c_id] = federated_clients[c_id].concatenate(data)
    return federated_clients

# print(f"client_ids: {client_ids}", dataset)
client_ids = [ str(x) for x in np.arange(NUM_CLIENTS) ]
federated_clients = get_federated_clients(df, split_method="per_class")

def serializable_dataset_fn(client_id):
    return federated_clients[int(client_id)]

def get_client_data(client_ids):
    return tff.simulation.datasets.ClientData.from_clients_and_tf_fn(
        client_ids,
        serializable_dataset_fn
    )

def make_federated_data(client_data, client_ids):
    return [
        preprocess(client_data.create_tf_dataset_for_client(x))
        for x in client_ids
    ]

client_data = get_client_data(client_ids)
federated_data = make_federated_data(client_data, client_ids)
federated_data
```
